```python
from PIL import Image
import json
from javiface import JaviFace
from javiface import RetinaFace
```

```python
rf = RetinaFace(model_path = './javiface/artifacts/retinaface_r34.onnx', conf_threshold = 0.5)
```

```python
jf = JaviFace(onnx_path = './javiface/artifacts/javi_face_v1.onnx')
```

#### MetaData Modelo Javi Face V1

```python
with open('./javiface/artifacts/javi_face_v1_metadata.json', 'rb') as file:
    metadata = json.load(file)
```

```python
metadata
```

```python
img_1 = Image.open('./form_selfie_ejemplo.png')
img_2 = Image.open('./example_rotation.png')
```

```python
face_1, conf = rf.crop_face_rf(img_1, vis_threshold = 0.8, expand_face_area=0.2)
face_2, conf = rf.crop_face_rf(img_2, vis_threshold = 0.8, expand_face_area=0.2)
```

```python
results = jf.compare(img_1, img_2, threshold = metadata['threshold_selfie_vs_selfie'])
results
```
